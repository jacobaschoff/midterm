```python
import pandas as pd
import numpy as np

china = pd.read_csv('https://raw.githubusercontent.com/jacobaschoff/midterm/main/China%20Dataset.csv')
fentanyl1 = "Heterocyclic compounds; containing an unfused pyridine ring (whether or not hydrogenated) in the structure, alfentanil (INN), anileridine (INN), bezitramide (INN), bromazepam (INN), carfentanil (INN), difenoxin (INN), diphenoxylate (INN), dipipanone (INN)"
fentanyl2 = "Heterocyclic compounds; containing an unfused pyridine ring (whether or not hydrogenated) in the structure, alfentanil (INN), anileridine (INN), bezitramide (INN), bromazepam (INN), carfentanil (INN), difenoxin (INN), diphenoxylate (INN), dipipanone (INN)"

```

```python
china
```

```python
chinaCommodity1 = china[china['Commodity'] == fentanyl1]
chinaCommodity2 = china[china['Commodity'] == fentanyl2]
chinaconcat = [chinaCommodity1, chinaCommodity2]
chinaCommodity = pd.concat(chinaconcat)

```

```python
chinaCommodity
```

```python
chinaExports = chinaCommodity[chinaCommodity['Flow'] == 'Export']
chinaImports = chinaCommodity[chinaCommodity['Flow'] == 'Import']
chinaWeight = pd.DataFrame(chinaExports['Weight (kg)'])
chinaWeight.sum(0)
```
